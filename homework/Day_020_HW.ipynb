```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件
plt.style.use('ggplot')

# 忽略警告訊息
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

### 作業1:
請用 numpy 建立一個 10 x 10, 數值分布自 -1.0 ~ 1.0 的矩陣並繪製 Heatmap

```python
"""
Your Code Here
"""
matrix = 2*np.random.randn(10, 10)
#將大於1的element變成1，將小於-1的element變成-1
matrix[matrix > 1] = 1
matrix[matrix < -1] = -1

plt.figure(figsize=(10,10))
ax = sns.heatmap(matrix, cmap = plt.cm.RdYlBu_r, vmin = -1, annot = True, vmax = 1)
ax.set_ylim(len(matrix), 0) #為了防止 第一個row 與最後一個row 被砍一半的Bug
plt.title('Heatmap');
plt.show()
# heatmap = 
# plt.show()
```

### 作業2:
請用 numpy 建立一個 1000 x 3, 數值分布為 -1.0 ~ 1.0 的矩陣，並繪製 PairPlot (上半部為 scatter, 對角線為 hist, 下半部為 density)

```python
nrow = 1000
ncol = 3


matrix = np.random.randn(nrow, ncol)
#將大於1的element變成1，將小於-1的element變成-1
matrix[matrix > 1] = 1
matrix[matrix < -1] = -1 

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice).reset_index()

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False, 
                    hue = 'index', vars = [x for x in list(plot_data.columns) if x != 'index'])

#上半部為 scatter
grid.map_upper(plt.scatter, alpha = 0.2)

# 對角線畫 histogram
grid.map_diag(plt.hist, edgecolor="w")

# 下半部放 density plot
grid.map_lower(sns.kdeplot, cmap = plt.cm.OrRd_r, shade=True)


#加上 legend
grid = grid.add_legend()
plt.show()
```

### 作業3:
請用 numpy 建立一個 1000 x 3, 數值分布為常態分佈的矩陣，並繪製 PairPlot (上半部為 scatter, 對角線為 hist, 下半部為 density)

```python
nrow = 1000
ncol = 3


matrix = np.random.randn(nrow, ncol)
# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice(['one', 'two', 'three'], size=nrow)
plot_data = pd.DataFrame(matrix, indice).reset_index()

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False,
                    hue = 'index', vars = [x for x in list(plot_data.columns) if x != 'index'])

grid.map_upper(plt.scatter, alpha = 0.2)
grid.map_diag(plt.hist, edgecolor="w")
grid.map_lower(sns.kdeplot ,cmap = plt.cm.OrRd_r, shade=True)

#加上 legend
grid = grid.add_legend()

plt.show()
```
