```python
import pandas as pd
from sklearn.metrics import accuracy_score 
from sklearn.model_selection import cross_val_score
import numpy as np
```

```python
train_data = pd.read_csv('./data-science-london-scikit-learn/train.csv',header = None)
train_labels = pd.read_csv('./data-science-london-scikit-learn/trainLabels.csv',header = None)
test_data =  pd.read_csv('./data-science-london-scikit-learn/test.csv',header = None)
```

```python
print(train_data.shape)
print(train_labels.shape)
print(test_data.shape)
```

```python
train_data.describe()
```

### PRE-PROCESSING
* Train-Test Split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(train_data, train_labels, test_size = 0.30, random_state = 101)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### without feature scaling

```python
#LOGISTIC REGRESSION
from sklearn.linear_model import LogisticRegression

lr_model = LogisticRegression(solver = 'saga')
lr_model.fit(x_train, y_train.values.ravel())
lr_predicted = lr_model.predict(x_test)
print('Accuracy = ',accuracy_score(y_test, lr_predicted))
```

```python
# Cross validation
scores = cross_val_score(lr_model, train_data, train_labels.values.ravel(), cv=5, scoring='accuracy')
print('Accuracy = ', scores.mean())
```

### predict test data

```python
lr_test_predicted = lr_model.predict(test_data)
```

### 產生submission DataFrame

```python
import numpy as np
submission_df = pd.DataFrame()
submission_df['Id'] = np.arange(test_data.shape[0])+1
submission_df['Solution'] = lr_test_predicted
submission_df
```

### 存成 csv

```python
#產生submission DataFrame
submission_df.to_csv('submission_no_normalization.csv', index=False)
```

### feature scaling

```python
df = pd.concat([train_data, test_data]) #上下合併成一個大的DataFrame (方便做 Scaling)
df
```

### 三種不同的 Scaling

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, Normalizer

std = StandardScaler()
df_std = std.fit_transform(df) #對每一個 column 轉換
mms = MinMaxScaler()
df_mms = mms.fit_transform(df) #對每一個 column 轉換
norm = Normalizer()
df_norm = norm.fit_transform(df) #對每一個 sample 轉換 (2 norm = 1)
```

```python
#將 Scaling 後的資料拆回 train 與 test
train_data_std = df_std[:1000]
test_data_std = df_std[1000:]

train_data_mms = df_mms[:1000]
test_data_mms = df_mms[1000:]

train_data_norm = df_norm[:1000]
test_data_norm = df_norm[1000:]
```

```python
x_train_std, x_test_std, y_train, y_test = train_test_split(train_data_std, train_labels, test_size = 0.30, random_state = 101)

lr_model_std = LogisticRegression(solver = 'saga')
lr_model_std.fit(x_train_std, y_train.values.ravel())
lr_predicted_std = lr_model_std.predict(x_test_std)
print('Accuracy = ',accuracy_score(y_test, lr_predicted_std))
```

```python
# Cross validation
scores = cross_val_score(lr_model_std, train_data_std, train_labels.values.ravel(), cv=5, scoring='accuracy')
print('Accuracy = ', scores.mean())
```

```python
x_train_mms, x_test_mms, y_train, y_test = train_test_split(train_data_mms, train_labels, test_size = 0.30, random_state = 101)

lr_model_mms = LogisticRegression(solver = 'lbfgs')
lr_model_mms.fit(x_train_mms, y_train.values.ravel())
lr_predicted_mms = lr_model_mms.predict(x_test_mms)
print('Accuracy = ',accuracy_score(y_test, lr_predicted_mms))
```

```python
# Cross validation
scores = cross_val_score(lr_model_mms, train_data_mms, train_labels.values.ravel(), cv=5, scoring='accuracy')
print('Accuracy = ', scores.mean())
```

```python
x_train_norm, x_test_norm, y_train, y_test = train_test_split(train_data_norm, train_labels, test_size = 0.30, random_state = 101)

lr_model_norm = LogisticRegression(solver = 'saga')
lr_model_norm.fit(x_train_norm, y_train.values.ravel())
lr_predicted_norm = lr_model_norm.predict(x_test_norm)
print('Accuracy = ',accuracy_score(y_test, lr_predicted_norm))
```

```python
# Cross validation
scores = cross_val_score(lr_model_norm, train_data_norm, train_labels.values.ravel(), cv=5, scoring='accuracy')
print('Accuracy = ', scores.mean())
```

### predict testing data

```python
lr_test_predicted_std = lr_model_std.predict(test_data_std)
lr_test_predicted_mms = lr_model_std.predict(test_data_mms)
lr_test_predicted_norm = lr_model_std.predict(test_data_norm)
```

### 產生 三種 submission DataFrame

```python
# std
submission_df_std = pd.DataFrame()
submission_df_std['Id'] = np.arange(test_data.shape[0])+1
submission_df_std['Solution'] = lr_test_predicted

# mms
submission_df_mms = pd.DataFrame()
submission_df_mms['Id'] = np.arange(test_data.shape[0])+1
submission_df_mms['Solution'] = lr_test_predicted_mms

#norm
submission_df_norm = pd.DataFrame()
submission_df_norm['Id'] = np.arange(test_data.shape[0])+1
submission_df_norm['Solution'] = lr_test_predicted_norm
```
