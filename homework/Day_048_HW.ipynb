```python
import pandas as pd
from sklearn.metrics import accuracy_score 
from sklearn.model_selection import cross_val_score
```

```python
train_data = pd.read_csv('./data-science-london-scikit-learn/train.csv',header = None)
train_labels = pd.read_csv('./data-science-london-scikit-learn/trainLabels.csv',header = None)
test_data =  pd.read_csv('./data-science-london-scikit-learn/test.csv',header = None)
```

```python
print(train_data.shape)
print(train_labels.shape)
print(test_data.shape)
```

```python
train_data.describe()
```

### PRE-PROCESSING
* Train-Test Split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(train_data,train_labels, test_size = 0.30, random_state = 101)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#LOGISTIC REGRESSION
from sklearn.linear_model import LogisticRegression

lr_model = LogisticRegression(solver = 'saga')
lr_model.fit(x_train,y_train.values.ravel())
lr_predicted = lr_model.predict(x_test)
print('Logistic Regression',accuracy_score(y_test, lr_predicted))

#SVM
from sklearn.svm import SVC

svc_model = SVC(gamma = 'auto', probability=True)
svc_model.fit(x_train,y_train.values.ravel())
svc_predicted = svc_model.predict(x_test)
print('SVM',accuracy_score(y_test, svc_predicted))
```

### predict test data

```python
lr_test_predicted = lr_model.predict(test_data)
svc_test_predicted = svc_model.predict(test_data)
```

### 產生submission DataFrame

```python
import numpy as np
submission_df = pd.DataFrame()
submission_df['Id'] = np.arange(test_data.shape[0])+1
submission_df['Solution'] = lr_test_predicted
submission_df
```

```python
#產生submission DataFrame
submission_df.to_csv('submission_no_normalization.csv', index=False)
```
