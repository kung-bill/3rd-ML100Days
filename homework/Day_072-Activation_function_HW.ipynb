# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''
```

```python
# ReLU (Rectified Linear Unit,) 數學函數表示方式
#sigmoid = lambda x: max(0, x) # Lambd 表示法
def ReLU(x):
    return np.maximum(0, x)

#ReLU 微分
def dReLU(x):
    y = np.zeros(len(x))
    for i in range(len(x)):
        if x[i] > 0:
            y[i] = 1
        else:
            y[i] = 0
    return y
```

```python
#畫圖
x = plt.linspace(-5,5,100)

plt.subplot(2,1,1)
# prepare the plot, associate the color r(ed) or b(lue) and the label 
plt.plot(x, ReLU(x), 'b', label='linspace(-10,10,10)')

# Draw the grid line in background.
plt.grid()

# 顯現圖示的Title
plt.title('ReLU Function')

# 顯現 the Sigmoid formula
plt.text(-4, 3, r'$ReLU(x)= \max(0, x)$', fontsize=15)

#resize the X and Y axes
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
plt.gca().yaxis.set_major_locator(plt.MultipleLocator(1))


plt.subplot(2,1,2)
# prepare the plot, associate the color r(ed) or b(lue) and the label 
plt.plot(x, dReLU(x), 'b', label='linspace(-10,10,10)')

# Draw the grid line in background.
plt.grid()

# 顯現圖示的Title
plt.title('ReLU Function')

# 顯現 the Sigmoid formula
plt.text(-5.3, 0.5, r'$\frac{d}{dx} ReLU(x)= \mathcal{I}_{\{x>0\}}(x)$', fontsize=15)

#resize the X and Y axes
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
plt.gca().yaxis.set_major_locator(plt.MultipleLocator(0.1))
```
